```python
import numpy as np
```

```python
#defining sigmoidfunction
def sigmoid(x):
    return 1/(1+np.exp(-x))

#defining derivative of function
def sigmoid_derivative(x):
    return x*(1-x)
```

```python
#defining training data for XOR classifier
x=np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])  
```

```python
#defining training labels
y=np.array([[0],[1],[1],[0]])  #(4,1)
```

```python
#initializing weights and bias
np.random.seed(1)
input_size=2
hidden_state=4
output_size=1

w1 = np.random.randn(input_size,hidden_state)   
b1=np.zeros((1,hidden_state))                   
w2=np.random.randn(hidden_state,output_size)    
b2=np.zeros((1,output_size))                        

#training
learning_rate=0.1
epochs=100000

for _ in range(epochs):
    #forward pass
    z1 = np.dot(x,w1) + b1    
    a1=sigmoid(z1)            
    
    z2=np.dot(a1,w2) + b2    
    output=sigmoid(z2)       
    
    #loss function
    loss = np.mean((y-output)**2)
    
    #back propagation
    d_output = (y-output)*sigmoid_derivative(output)   
    
    d_w2 = np.dot(a1.T,d_output) 
    d_b2 = np.sum(d_output,axis=0,keepdims=True) 
    
    d_hidden = np.dot(d_output,w2.T)*sigmoid_derivative(a1) 
    
    d_w1 = np.dot(x.T,d_hidden)    
    d_b1 = np.sum(d_hidden,axis=0,keepdims=True)
    
    w2+= d_w2*learning_rate
    b2+= d_b2*learning_rate
    w1+= d_w1*learning_rate
    b1+= d_b1*learning_rate
    
#finding output
print(loss)
print("predicting values:")
for i in x:
    h=sigmoid(np.dot(i,w1)+b1)
    o=sigmoid(np.dot(h,w2)+b2)
    print(f"{i} -> {round(o[0][0])}")
```
